```python
# Convert column names to snake_case
df.columns = (
    df.columns
    .str.strip()  # Remove leading/trailing spaces
    .str.lower()  # Convert to lowercase
    .str.replace(r"[-/\s+]", "_", regex=True)  # Replace spaces and special characters with "_"
)
df.info()
```

<div style="padding:15px; border:3px solid #3d7ab3; border-radius:6px;">
    <strong>Categorical Labels</strong> <br>
    📌 Convert all categorical labels to snake_case for consistency, improved readability, and to minimize the risk of errors.
</div>

```python
def standardize_categorical_labels(categorical_label):
    return (
        categorical_label
        .strip()  # Remove leading/trailing spaces
        .lower()  # Convert to lowercase
        .replace("-", "_")  # Replace hyphens with "_"
        .replace("/", "_")  # Replace slashes with "_"
        .replace(" ", "_")  # Replace spaces with "_"
    )


# Define categorical columns to standardize labels
columns_to_standardize = ["heart_disease"]

# Apply standardization of categorical labels
for column in columns_to_standardize:
    df[column] = df[column].apply(standardize_categorical_labels)
```

```python
df['heart_disease'] = df['heart_disease'].map({"presence":1, "absence": 0})
```

```python
x = df.drop("heart_disease", axis=1)
y = df["heart_disease"]


x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42, stratify = y)

y_train.value_counts()
```

<div style="background-color:#3d7ab3; color:white; padding:12px; border-radius:6px;">
    <h2 style="margin:0px">pre-processing of train data</h2>
</div>


```python
#erasing bad predictors
x_train = x_train.drop([ "bp", "cholesterol", "fbs_over_120", "ekg_results"], axis=1)
x_test = x_test.drop([ "bp", "cholesterol", "fbs_over_120", "ekg_results"], axis=1)

#removing outliers
outlier_remove = OutlierRemoverIQR()
outlier_remove.fit(df = x_train, numerical_columns = ["age", "max_hr", "st_depression"])
x_train = outlier_remove.transform(x_train)
y_train = y_train.loc[x_train.index]
```

```python
#let's standardize train and test
sc = StandardScaler()
cols_to_scale = ["age", "max_hr", "st_depression"]
x_train[cols_to_scale] = sc.fit_transform(x_train[cols_to_scale])
x_test[cols_to_scale] = sc.transform(x_test[cols_to_scale])
```

```python
#makin a one-hot enconding to ordinal categorical columns
cols_to_encode = ["sex","chest_pain_type","exercise_angina", "slope_of_st", "number_of_vessels_fluro", "thallium"]

df_pandas_encoded = pd.get_dummies(df, columns=cols_to_encode, drop_first=True)

encoder = OneHotEncoder(sparse_output=False)

one_hot_encoded = encoder.fit_transform(x_train[cols_to_encode])
one_hot_df = pd.DataFrame(one_hot_encoded, 
                          columns=encoder.get_feature_names_out(cols_to_encode), 
                          index = x_train.index)

x_train_processed= pd.concat([x_train.drop(cols_to_encode, axis=1), one_hot_df], axis=1)

one_hot_encoded = encoder.transform(x_test[cols_to_encode])

one_hot_df = pd.DataFrame(one_hot_encoded, 
                          columns=encoder.get_feature_names_out(cols_to_encode), 
                          index = x_test.index)

x_test_processed = pd.concat([x_test.drop(cols_to_encode, axis=1), one_hot_df], axis=1)
```

```python
#Logistic regression
clf = LogisticRegression(max_iter=10000, random_state=42)

clf.fit(x_train_processed, y_train)

y_pred = clf.predict(x_test_processed)

print("--- Clasification Report ---")
print(classification_report(y_test, y_pred))

acc = accuracy_score(y_test, clf.predict(x_test_processed)) * 100
print("--- Accuracy report ---")
print(f"Logistic Regression model accuracy: {acc:.2f}%")
```

```python
#Confusion Matrix

cm = confusion_matrix(y_test, y_pred)

cm_percent = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]


labels = (np.array([f"{count}\n({percent:.2%})" 
          for count, percent in zip(cm.flatten(), cm_percent.flatten())]
         )).reshape(2, 2)

plt.figure(figsize=(8, 6))
sns.heatmap(cm_percent, annot=labels, fmt="", cmap='Greens', 
            xticklabels=['Predicho Sano', 'Predicho Enfermo'], 
            yticklabels=['Real Sano', 'Real Enfermo'])

plt.xlabel('Predicción')
plt.ylabel('Realidad')
plt.title('Matriz de Confusión: Porcentajes de Acierto por Clase')
plt.show()
```
